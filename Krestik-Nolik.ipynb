```python
def draw_board(board, n):
    print('----'*n + '-')
    for i in range(n):
        print('|', end=' ')
        for j in range(n):
            print(board[i*n+j], end=' ')
            if(j == n - 1):
                print('|')
            else:
                print('|', end=' ')
        print('----'*n + '-')

def draw_value(board, n, i, j, me): # me - who's move
    if(me):
        board[i*n+j] = 'x'
    else:
        print(i, j)
        board[i*n+j] = 'o'

def check_winner(board, n, k): # 0 - none, 1 - me, 2 - he
    # проверим есть ли победитель по горизонтали
    cnt_me = 0
    cnt_he = 0
    for i in range(len(board)):
        if(i % n == 0):
            cnt_me = 0
            cnt_he = 0
        if(board[i] == 'x'):
            cnt_me+=1
            cnt_he = 0
        elif(board[i] == 'o'):
            cnt_he+=1
            cnt_me = 0
        else:
            cnt_he = 0
            cnt_me = 0
        if(cnt_me == k):
            return 1
        if(cnt_he == k):
            return 2
    #проверим есть ли победитель по вертикали
    cnt_me = 0
    cnt_he = 0
    for s in range(n):
        for cur in range(n):
            if(board[s + cur*n] == 'x'):
                cnt_me+=1
                cnt_he = 0
            elif(board[s + cur*n] == 'o'):
                cnt_he+=1
                cnt_me = 0
            else:
                cnt_me = 0
                cnt_he = 0
            if(cnt_me == k):
                return 1
            if(cnt_he == k):
                return 2
        cnt_me = 0
        cnt_he = 0
    #проверим есть ли победитель по диагоналям
    ind = 0
    for i in range(n*2-1):
        cind = ind
        arr = []
        need = n - abs(i+1 - n)
        while cind >= 0 and len(arr) < need:
            arr.append(cind)
            cind -= n-1
        if i < n-1:
            ind += n
        else:
            ind += 1
        #в arr список индексов на текущей диагонали
        #print(arr)
        cnt_me = 0
        cnt_he = 0
        for pos in arr:
            if(board[pos] == 'x'):
                cnt_me+=1
                cnt_he = 0
            elif(board[pos] == 'o'):
                cnt_he+=1
                cnt_me = 0
            else:
                cnt_he = 0
                cnt_me = 0
            if(cnt_me == k):
                return 1
            if(cnt_he == k):
                return 2

    ind = n-1
    for i in range(n*2-1):
        cind = ind
        arr = []
        need = n - abs(i+1 - n)
        while cind >= 0 and len(arr) < need:
            arr.append(cind)
            cind -= n + 1
        if i < n-1:
            ind += n
        else:
            ind -= 1
        #в arr список индексов на текущей диагонали
        #print(arr)
        cnt_me = 0
        cnt_he = 0
        for pos in arr:
            if(board[pos] == 'x'):
                cnt_me+=1
                cnt_he = 0
            elif(board[pos] == 'o'):
                cnt_he+=1
                cnt_me = 0
            else:
                cnt_he = 0
                cnt_me = 0
            if(cnt_me == k):
                return 1
            if(cnt_he == k):
                return 2
    return 0
```

```python
message = "Send me N, please.\n"
n = int(input(message))
message = "Send me K, please.\n"
k = int(input(message))
```

```python
board = [' '] * n * n
have = []
for i in range(1, n + 1):
    for j in range(1, n + 1):
        cur = []
        cur.append(i)
        cur.append(j)
        have.append(cur)
```

```python
cnt = 0
who_win = 0
while(True):
    who_win = check_winner(board, n, k)
    if(who_win != 0):
        break
    if(cnt % 2 == 0): #мой ход
        draw_board(board, n)
        message = "Send me i, please."
        curi = int(input(message))
        message = "Send me j, please."
        curj = int(input(message))
        draw_value(board, n, curi - 1, curj - 1, 1)
        have.remove([curi, curj])
    else:
        cur = random.choice(have)
        curi = cur[0]
        curj = cur[1]
        draw_value(board, n, curi - 1, curj - 1, 0)
        have.remove([curi, curj])
    cnt+=1

print("\nCongratulations player {}".format(who_win))
draw_board(board, n)
```
